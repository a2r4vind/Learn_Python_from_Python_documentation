# 4.8 Defining Functions

```python
def fib(n):
    ''' return a list of fibonacci series upto n'''
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)
        a, b = b, a+b
    return result

f100 = fib(100)
f100
    
```

## 4.9.1 Default Argument Values

```python
i = 5 
def f(arg = i):
    print(arg)

i = 6
f()
```

```python
def f(a, l = []):
    l.append(a)
    return l

print(f(1))
print(f(2))
print(f(3))
```

```python
def fu(a, l=None):
    if l is None:
        l = []
        
    l.append(a)
    return l

print(fu(1))
print(fu(2))
print(fu(3))
```

## 4.9.2 Keyword Arguments

***args** always comes before ****kwargs**

*args => tuple

**kwargs => dictionary

```python
def cheeseshop(kind, *args, **kwargs):
    print(f"--Do you have any {kind}?")
    print(f"--I'm sorry, we're all out of {kind}.")
    for arg in args:
        print(arg)
    print("-" * 40)
    for kw in kwargs:
        print(kw, ":", kwargs[kw])
```

```python
cheeseshop('Hamburger',"It's very runny, sir.", "It's really very, very runny, sir.",
          shopkeeper="Arjun",
          client="Abhinav",
          sketch="Cheese Shop Sketch")
```

## 4.9.3 Special Parameters

### 4.9.3.1 Positional or Keyword Arguments

--> If **(/)** or **(*)** are not present in function definition then arguments may be passed to a function by position or by keyword

### 4.9.3.2 Positional Only Parameters
--> Positional only parameters are placed before a **forward-slash(/)**.

--> **/** is used to logically separate the positional-only parameters from the rest of the parameters.

--> Parameter's order matters in positional-only parameters.

--> positional-only parameters cannot be passed by keyword.

### 4.9.3.3 Keyword Only Parameters
--> place **(*)** before first keyword arguments.

--> pass parameter by keyword.

```python
def standard_arg(arg):
    print(arg)

def pos_only_arg(arg, /):
    print(arg)

def kwd_only_arg(*, arg):
    print(arg)

def combined_example(pos_only, /, standard, *, kwd_only):
    print(f"Position Only parameter {pos_only}, Standard parameter {standard}, Keyword Only parameter {kwd_only}")

```

```python
standard_arg(2) # position
```

```python
standard_arg(arg=2) # keyword
```

```python
pos_only_arg(4)
```

```python
pos_only_arg(arg=4)
```

```python
kwd_only_arg(arg=6)
```

```python
kwd_only_arg(6)
```

```python
combined_example(4, 2, kwd_only=6)
```

```python
combined_example(4, standard=2, kwd_only=6)
```

## 4.9.4 Arbitrary ( unspecified) argument list
--> A function can be called with arbitrary number of arguments which are wrapped up in a tuple as *args.

--> Before this *args any number of formal parameters can be present either zero or more.

--> After this *args any formal parameter can be used as keyword only not as positional.

```python
def write_multiple_items(file, separator, *args):
    file.write(separator.join(args))
```

```python
def concat(*args, sep="/"):
    print(sep.join(args))

concat('Arvind','is','a','good','learner')
```

```python
concat('Arvind','is','a','good','learner',sep=',')
```

## 4.9.6 Lambda Expression
--> Used to create small anonymous functions with **lambda** keyword.

--> Mostly used wherever function objects are required.

```python
def make_incrementor(n): # There is need to create object of this function
    return lambda x: x + n

func_obj = make_incrementor(100)
```

```python
func_obj(1)
```

```python
func_obj(21)
```

--> lambda functions can be used to pass a small function as an argument

```python
pairs = [(1,'one'),(2,'two'),(3,'three'),(4,'four')]
pairs.sort(key=lambda pair: pair[1])
print(pairs)
```

## 4.9.8 Function Annotations

```python
def f(ham: str, eggs: str = 'eggs') -> str:
    print("Annotations are :", f.__annotations__)
    print("Arguments : ", ham, eggs)
    return ham + ' and ' + eggs

f('spam')
```

--> Annotations are stored in the __ annotations __ attribute of the function as the dictionary.

--> Annotation parameters are defined by colon ( : ) after the parameter name followed by the expression evaluating to the value of the annotation.

--> Return annotations are defined by literal (->) followed by the expression.

--> End of the def is defined by colon (:)

## 4.9.9 Intermezzo: Coding Style

1) Instead of using TAB; use 4 space indentation
2) Wrap line so that they don't exceed 79 characters.
3) Use Docstrings
4) Whenever possible put comments on a line
5) Leave a blank line to separate functions, methods and large bolck of codes inside functions.
6) Use space around operators and after the commas, but not directly inside the bracketing constructs. eg. a = f(1, 2) + g(4, 8)
7) Use UpperCamleCase for classes and lowercase_with_underscores for functions and methods.
8) Don't use fancy encodings or non-ASCII characters in Identifiers.
