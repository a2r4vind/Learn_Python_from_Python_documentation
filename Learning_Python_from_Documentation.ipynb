# 4.8 Defining Functions

```python
def fib(n):
    ''' return a list of fibonacci series upto n'''
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)
        a, b = b, a+b
    return result

f100 = fib(100)
f100
    
```

## 4.9.2 Keyword Arguments

***args** always comes before ****kwargs**

*args => tuple

**kwargs => dictionary

```python
def cheeseshop(kind, *args, **kwargs):
    print(f"--Do you have any {kind}?")
    print(f"--I'm sorry, we're all out of {kind}.")
    for arg in args:
        print(arg)
    print("-" * 40)
    for kw in kwargs:
        print(kw, ":", kwargs[kw])
```

```python
cheeseshop('Hamburger',"It's very runny, sir.", "It's really very, very runny, sir.",
          shopkeeper="Arjun",
          client="Abhinav",
          sketch="Cheese Shop Sketch")
```

## 4.9.3 Special Parameters

### 4.9.3.1 Positional or Keyword Arguments

--> If **(/)** or **(*)** are not present in function definition then arguments may be passed to a function by position or by keyword

### 4.9.3.2 Positional Only Parameters
--> Positional only parameters are placed before a **forward-slash(/)**.

--> **/** is used to logically separate the positional-only parameters from the rest of the parameters.

--> Parameter's order matters in positional-only parameters.

--> positional-only parameters cannot be passed by keyword.

### 4.9.3.3 Keyword Only Parameters
--> place **(*)** before first keyword arguments.

--> pass parameter by keyword.

```python
def standard_arg(arg):
    print(arg)

def pos_only_arg(arg, /):
    print(arg)

def kwd_only_arg(*, arg):
    print(arg)

def combined_example(pos_only, /, standard, *, kwd_only):
    print(f"Position Only parameter {pos_only}, Standard parameter {standard}, Keyword Only parameter {kwd_only}")

```

```python
standard_arg(2) # position
```

```python
standard_arg(arg=2) # keyword
```

```python
pos_only_arg(4)
```

```python
pos_only_arg(arg=4)
```

```python
kwd_only_arg(arg=6)
```

```python
kwd_only_arg(6)
```

```python
combined_example(4, 2, kwd_only=6)
```

```python
combined_example(4, standard=2, kwd_only=6)
```

## 4.9.8 Function Annotations

```python
def f(ham: str, eggs: str = 'eggs') -> str:
    print("Annotations are :", f.__annotations__)
    print("Arguments : ", ham, eggs)
    return ham + ' and ' + eggs

f('spam')
```

--> Annotations are stored in the __ annotations __ attribute of the function as the dictionary.

--> Annotation parameters are defined by colon ( : ) after the parameter name followed by the expression evaluating to the value of the annotation.

--> Return annotations are defined by literal (->) followed by the expression.

--> End of the def is defined by colon (:)

### 5.1.3 List Comprehension

--> When to use:

2) To create a subsequence of those elements from a sequence that satisfy a certain condition.

```python
# create list of squares of natural numbers
n = int(input('Enter a natural number : '))
squares = [x**2 for x in range(1,n+1)]
print(squares)
```

## 5.6 Looping techniques

```python
# In dictionary the key and corresponding value can be retrieve at the same time using items() method.
for fruit, color in fruits_colors_dict.items():
    print(f"{fruit} => {color}")
```

```python
# when looping through a sequence like list, tuple, etc. the position index and corresponding value 
# can be retrieved at the same time using enumerate() function.

seq = ['apple', 'banana', 'grapes', 'kiwi', 'dragon fruit', 'peach']
for idx, fruit in enumerate(seq):
    print(f"{idx} => {fruit}")
```

```python
# To loop over two or more sequences at the same time, zip() function can be used.

for fruit, color in list(zip(fruits, colors)):
    print(f'{fruit} => {color}')
```

```python
# To loop over a sequence in reversed order use reversed() function
nums = [3,2,5,6,9,7,1,0]
for num in reversed(nums):
    print(num)
```

```python
# To loop over a sequence in sorted order use sorted() function
for num in sorted(nums):
    print(num)
```

```python
# To loop over unique elements of a sequence in sorted order use sorted() with set().
numbers = [8,3,2,5,2,7,8,9,9,2,1]
for num in sorted(set(numbers)):
    print(num)
```

## 8.3 Handling Exceptions

```python
while True:
    try:
        x = int(input("Enter a number: "))
        break
    except ValueError:
        print('Oops! That was no valid number. Try again...')
```

**Working of try...except**
1) Firstly, the **try** clause( the statement(s) between **try** and **except** keywords) is executed.
2) If no exception occurs, then **except** clause is skipped and execution of **try** statement is finished.
3) if exception occurs in any of the statements of the **try** clause during execution, then  remaining statements after that statement which caused an exception are skipped.
4) Then exception type matches with the **exception named** after the **except** keyword, if matched then **except** clause is executed followed by the rest of the code after the **try...except block**.
5) If **exception type** doesn't matches with the **exception named** in **except clause** then it is passed on to outer **try** statements; if no handler is present; then it is an unhandled exception  and execution is stopped with an error message.

**Note:-** 
1) A **try** statement may have more than one **except** clause, to specify handlers for different exceptions.
2) At most one handler will be executed.
3) Handlers handle exceptions that occur in the corresponding ***try*** clause.
4) An except clause may name multiple exceptions.

        >>> except RuntimeError, TypeError, NameError:
                pass

1) The except clause may specify variable after the exception name.
2) That variable is bound to exception instance and has **args** attribute which stores the arguments.

```python
try:
    raise Exception('spam', 'eggs')
except Exception as inst:
    print(type(inst))
    print(inst.args)
    print(inst)
    x, y = inst.args
    print(f'x = {x}')
    print(f'y = {y}')
```

**try...except clause** also has **else clause** which is executed if **try clause** doesn't raise an exception.

```python
try:
    l = [1,3,4,5,2,6]
    print(l[9])
except IndexError as err:
    print(err)
else:
    print('Finished!')
```

```python
try:
    l = [1,3,4,5,2,6]
    print(l[4])
except IndexError as err:
    print(err)
else:
    print('Finished!')
```

An exception which is direct cause of another can be chained using optional **from** clause of **raise** statement

    >>> raise exception_type from exc
    where exc must be exception_instance or None

```python
def func():
    raise ConnectionError
    
try:
    func()
except ConnectionError as exc:
    raise RuntimeError('Failed to open database') from exc
```

## 8.7 Defining Clean-up Actions
Uses **finally** clause

```python
def divide(x,y):
    try:
        result = x / y
    except ZeroDivisionError:
        print('Division by Zero!')
    else:
        print(f'The result of {x} / {y} = {result}')
    finally:
        print('executing the finally clause.')
```

```python
divide(10,2)
```

```python
divide(10,0)
```

## 9.2 Python Scopes and Namespaces
1) **Namespace** is a mapping from names to objects.
2) Namespaces are created at different moments and have different lifetimes. Eg. Namespace for built-in names is created when Python interpreter starts up and is never deleted. Another example is global namespace for a module is created when the module definition is read and it last untill the interpreter quits. And local namespace is created when the function is called and last untill the function returns or raises an exception which is not handle within the function.
3) **Scope** is textual region of a Python program where a namespace is directly accessible.


```python
def scope_test():
    def do_local():
        spam = 'local spam'
        
    def do_nonlocal():
        nonlocal spam
        spam = 'nonlocal spam'
        
    def do_global():
        global spam
        spam = 'global spam'
        
    spam = 'test spam'
    do_local()
    print('After local assignment:', spam) # test spam
    do_nonlocal()
    print('After nonlocal assignment:', spam) # nonlocal spam
    do_global()
    print('After global assignment:', spam) # nonlocal spam
    
scope_test()
print('In global scope:', spam) # global spam
```

**Note:-** Class variables must not be mutable objects like list, etc

```python
# tricks list should not be used in the following code as a class variable because 
# that single list would be shared by all Dog instances.

class Dog:
    """A dog class."""
    
    tricks = [] # This must not be used as class variable
    
    def __init__(self, name):
        self.name = name 
        
    def add_trick(self, trick):
        self.tricks.append(trick)
        
d = Dog('Buddy')
e = Dog('Leo')

d.add_trick('roll over')
e.add_trick('play dead')

d.tricks # unexpectedly shared by all dogs
```

```python
# tricks list should be specific to individual dog 
class Dog:
    """A dog class."""
        
    def __init__(self, name):
        self.name = name 
        self.tricks = [] 
        
    def add_trick(self, trick):
        self.tricks.append(trick)
        
d = Dog('Buddy')
e = Dog('Leo')

d.add_trick('roll over')
e.add_trick('play dead')

print(d.tricks)
print(e.tricks)
```

## 9.9 Generators
1) **Generators** are simple and powerful tools for creating iterators.
2) They are defined like regular functions but uses **yield** whenever coder wants to return data.
3) Whenever **generator** is created **__ iter__()** and **__ next__()** methods are created automatically.

```python
def reverse(data):
    for index in range(len(data) - 1, -1, -1):
        yield data[index]
```

```python
for char in reverse('Arvind'):
    print(char)
```
